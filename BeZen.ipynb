## 3. Correct Product price in correct product format

### converting the string type price_string to float type

```python
def clean_currency(x):
    """ If the value is a string, then remove currency symbol and delimiters
    otherwise, the value is numeric and can be converted
    """
    if isinstance(x, str):
        return(x.replace('$', '').replace(',', ''))
    return(x)
```

```python
d2['price_string'] = d2['price_string'].apply(clean_currency).astype('float')
d2.head()
```

### again converting price_string to string to get the correct format and then splitting it into two new columns value and currency

```python
def currency(x):
    """ If the value is a float, then it add the currency symbol
    and upto 2 decimal places are added.
    """
    if isinstance(x, float):
        return("${:,.2f}".format(x))
    return(x)
```

```python
d2['price_string'] = d2['price_string'].apply(currency).astype('str')
d2.head()
```

```python
d2['value'] = d2['price_string'].apply(clean_currency).astype('float')
d2['currency']="$"
del d2['price_string']
d2.head()
```

## 4. List out categories with average pricee of the product

```python
d3= d2.groupby(['category']).mean()
print(d3)
plt.figure(figsize=(25,6))
plt.plot(d3.index,d3)
```
